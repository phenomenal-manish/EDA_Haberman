## Objective:
**To predict whether a patient will survive after 5 years or not based upon the patient's age,his/her operation_year and the number of positive axillary nodes .**

## BI-VARIATE ANALYSIS

## 2-D Scatter Plot

```python
# 2-D Scatter plot with color-coding for each survival status type/class.
# 'sns' corresponds to seaborn.
sns.set_style("whitegrid");
sns.FacetGrid(haberman_df, hue="Surv_status", height=5) \
   .map(plt.scatter, "Age", "axil_nodes") \
   .add_legend();
plt.show();
```

**Observation(s):**
1. Most of the patients have 0 positive axillary nodes detected.
2. We cannot seperate out blue points and orange points.Therefore, we cannot predict whether a patient would survive less than or more than 5 years using the above scatterplot(Age vs axil_nodes).
3. It is imperative to check all combinations/pairs of features to make a good classification or predicition.
4. Number of Combinations of features : 3C2 = 3 (excluding class-attribute 'Surv_status')
5. We need to plot Pair plots to analyse scatterplot combinations of all features.

## Pair Plot

```python
plt.close();
sns.set_style("whitegrid");
sns.pairplot(haberman_df, hue="Surv_status",
             vars=['Age','Op_Year','axil_nodes'], height=3)
plt.show()
#The diagnol elements are PDFs for each feature.
#each row has the y axis as mentioned   
#each column has the x-axis as mentioned
```

**Observation(s):**
1. axil_nodes vs Age seems to be an useful plot just to get an insight of the data. Most of the blue points are 0 which means that most of the people with 0 axil_nodes lived more than 5 years.
2. It seems that we cannot distinguish the data easily with the help of scalar plots of all combinations of features as all the scatter plots do not have separable data and the points are overlapping in each scatter plot.

## UNIVARIATE ANALYSIS

## Histogram, PDF

```python
#Histogram,PDF for Age
sns.FacetGrid(haberman_df, hue="Surv_status", height=5) \
   .map(sns.distplot, "Age") \
   .add_legend();
plt.show();
```

```python
#Histogram,PDF for Op_Year
sns.FacetGrid(haberman_df, hue="Surv_status", height=5) \
   .map(sns.distplot, "Op_Year") \
   .add_legend();
plt.show();
```

```python
#Histogram,PDF for axil_nodes
sns.FacetGrid(haberman_df, hue="Surv_status", height=5) \
   .map(sns.distplot, "axil_nodes") \
   .add_legend();
plt.show();
```

**Observation:**
1. Age and Op_Year are not good features to draw useful insight out of the data as the distibutions for both types of people who survived and did not survive almost overlap.The distributions are similar.

2. axil_nodes is a useful feature to know about the survival status of patients because there is difference between the distributions for both classes/types(labels). By looking at axil_nodes distibution we can infer that most patients who have survived for more than five years have 0 axil_nodes.

## CDF

```python
#divide the data set into two parts according to the label Surv_status 
#surv_more_5_years is for Surv_status=1
#died_within_5_years is for Surv_status =2
surv_more_5_years = haberman_df.loc[haberman_df["Surv_status"]==1]
died_within_5_years = haberman_df.loc[haberman_df["Surv_status"]==2]
```

```python
counts, bin_edges = np.histogram(surv_more_5_years['axil_nodes'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)
plt.legend(['Pdf for the patient who survived 5 years or longer',
            'Cdf for the patient who survived 5 years or longer'])
plt.show()
```

```python
counts, bin_edges = np.histogram(died_within_5_years['axil_nodes'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)
plt.legend(['Pdf for the patient who died within 5 year',
            'Cdf for the patient who died within 5 year'])
plt.show()
```

**Observation(s):**
1. From the above CDF of axil_nodes for the patients who survived 5 years or longer it is clear that more than 80% or roughly 90% of the people who survived have less than 10 axil_nodes.

2. By observing CDF of patients who died within 5 year, we can infer that patients above 47 axil_nodes will eventually die within 5 years. 

## Mean, Variance and Std-dev

```python
print("Summary Statistics of patients who survived 5 years or longer - ")
surv_more_5_years.describe()
```

```python
print("Summary Statistics of patients who died within 5 year - ")
died_within_5_years.describe()
```

**Observations:**
1. By looking at the summary statistics for both types we can observe that the statistics are similar for Age and Op_Year.

2. The values mean,std-deviation and percentiles of axil_nodes are quite different for the two types of patients.The mean of axil_nodes is more for people who died within 5 years than people who have survived who survived 5 years or longer.

## Median, Percentile, Quantile, IQR, MAD

```python
#Median, Quantiles, Percentiles, IQR.
print("\nMedians:")
print(np.median(surv_more_5_years["axil_nodes"]))
print(np.median(died_within_5_years["axil_nodes"]))

print("\nQuantiles:")
print(np.percentile(surv_more_5_years["axil_nodes"],np.arange(0, 100, 25)))
print(np.percentile(died_within_5_years["axil_nodes"],np.arange(0, 100, 25)))

print("\n90th Percentiles:")
print(np.percentile(surv_more_5_years["axil_nodes"],90))
print(np.percentile(died_within_5_years["axil_nodes"],90))

from statsmodels import robust
print ("\nMedian Absolute Deviation")
print(robust.mad(surv_more_5_years["axil_nodes"]))
print(robust.mad(died_within_5_years["axil_nodes"]))
```

## Box plot and Whiskers

```python
#Box-plot with whiskers: Method of visualizing 1-D scatter plot more intuitivey.

"""
Box plot takes less space and visually represents the five number summary of the data points in a box. 

1. Min value or Q1 - 1.5*IQR
2. Q1 (25th percentile)
3. Q2 (50th percentile or median)
4. Q3 (75th percentile)
5. Max value or Q3 + 1.5*IQR

- Inter Quartile Range = Q3 -Q1

- The outliers are displayed as points outside the box.

"""

sns.boxplot(x='Surv_status',y='axil_nodes', data=haberman_df)
plt.show()
```

```python
sns.boxplot(x='Surv_status',y='Age', data=haberman_df)
plt.show()
```

```python
sns.boxplot(x='Surv_status',y='Op_Year', data=haberman_df)
plt.show()
```

**Observation(s):**
1.By observing Box plot for axil_nodes, we can infer that most of the patients who survived more than 5 years have values for axil_nodes less than 4 or 5. 


## Violin plots

```python
# A violin plot combines the benefits of pdf's and whiskers box plot and simplifies them

# Denser regions of the data are fatter, and sparser ones thinner in a violin plot

sns.violinplot( x='Surv_status', y='Age', data=haberman_df)
plt.show()
```

```python
sns.violinplot( x='Surv_status', y='Op_Year', data=haberman_df)
plt.show()
```

```python
sns.violinplot( x='Surv_status', y='axil_nodes', data=haberman_df)
plt.show()
```

**Observation(s):**

4. Similarly, more no of patients who are dead had the operation between 1959-1965 and the patients who survived for more than 5 years had the operation between 1960-1966.
